```python
import pandas as pd
```

```python
data = pd.read_excel('Data Source.xlsx')
data
```

```python
data.isnull().sum()
```

```python
data.columns = ['Timestamp','Tipe Kendaraan','Jenis Kendaraan','Merk','Business','Prices/day']
```

```python
data = data[['Tipe Kendaraan','Jenis Kendaraan','Merk','Business','Prices/day']]
data
```

```python
data[['Jenis Kendaraan','Merk']] = data[['Jenis Kendaraan','Merk']].applymap(lambda x: x.lower())
data = data.applymap(lambda x: x.strip())
```

```python
data['Jenis Kendaraan'] = data['Jenis Kendaraan'].replace({
    'bebek, matic': 'bebek',
    'bebek, matik, sport': 'bebek',
    'motor bebek': 'bebek',
    'matik': 'matic',
    'metic': 'matic',
    'motor : matic': 'matic',
    'motor matic': 'matic',
    'motor: beat': 'matic',
    'motor: matic': 'matic',
    'motor:matic': 'matic',
    'minivan': 'minibus',
    'carcity': 'city car',
    'mpv':'minibus',
    'jeep': 'sport',
    'touring': 'sport',
    'rx king':'sport'
})
data['Jenis Kendaraan'] = data['Jenis Kendaraan'].replace({
    'minibus':'mpv',
    'sport':'suv',
    'city car':'hatchback'
})
```

```python
data.groupby(['Jenis Kendaraan'])['Business'].count().sort_values(ascending=False)
```

```python
data['Merk'] = data['Merk'].replace({
    'honda dan yamaha':'yamaha',
    'yamaha & honda':'yamaha'
})
```

```python
data.groupby(['Merk'])['Business'].count().sort_values(ascending=False)
```

```python
data['ID Num'] = range(1, len(data)+1)
data
```

```python
def create_file(x,name_file):
    writer = pd.ExcelWriter(name_file + '.xlsx')
    # write dataframe to excel
    x.to_excel(writer)
    # save the excel
    writer.save()
    print('DataFrame is written successfully to Excel File.')
```

```python
create_file(data, 'Data_Test')
```

```python
data_export = data.drop(columns='Merk').copy()
num_data = 1000
multiplier = num_data // len(data_export)+1
data_export = pd.concat([data_export] * multiplier, ignore_index=True)
data_export = data_export.sample(frac=1).reset_index(drop=True)
data_export
```

```python
create_file(data_export, 'Data_Training')
```
